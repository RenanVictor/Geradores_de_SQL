```python
def convert_str_to_float(dataframe,coluna):
    lista_numeric = []
    lista = dataframe[coluna].values.tolist()
    for item in lista:
        item = item.replace('.','')
        item = item.replace(',','.')
        item = item.replace(' kg','')
        item = item.replace(' %','')
        lista_numeric.append(float(item))
    dataframe[coluna] = lista_numeric
```

```python
def converter_tempo(texto):
    try:
        ponto = texto.find(":")
        hora = int(texto[:ponto])
        minuto = int(texto[ponto+1:])
        tempo =hora + round(minuto/60,2)

        return tempo
    except:
        return 00.00

```

```python
def conectarBD():  # Se conecta no servidor
    conexao = psy.connect(
        host="postgres.ata",
        database="pedidos",
        port=5000,
        user="renan",
        password="rev@123"
    )
    return conexao
    
def gera_cursor(conexao: psy):  # retorna o cursor
    cursor = conexao.cursor()
    return cursor
```

```python
def insert_bd(dataset: pd.DataFrame):
    sql = "INSERT INTO plan_laser(programa, espessura, maquina, inicio, Peso, aproveitamento, retalho, num_nesting, tempo, Tipo, status) VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s);"

    for item in dataset.values.tolist():
        cnx = conectarBD()
        cursor = gera_cursor(cnx)
        cursor.execute(sql, item)
        print(cursor.rowcount)
        cnx.commit()
```
